```python
%matplotlib inline

from modsim import *

population_table = System(Year = [1957, 1985, 1984, 1986, 1989, 1990, 1991, 1992, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002,
                             2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015],
           Population = [3491, 2176, 1961, 2087, 1761, 1524, 1685, 1495, 1508, 1496, 1473, 1509, 1503, 1539, 1343, 1308,
                          1261, 1386, 1166, 1214, 1132, 977, 1049, 1044, 1032, 1104, 1033, 920, 1034])

    
n0 = 1957
p0 = 3491

change = 1315
time = 28

slope = - change / time

results = TimeSeries()

results[n0] = p0

for t in linrange(1957, 1985):
    results[t+1] = results[t] + slope
    
alpha = 1
beta = -0.00002
    
for t in linrange(1986, 2015):
    results[t+1] = alpha * results[t] + beta * results[t]**2
    
plot(population_table.Year, population_table.Population, color='blue')
plot(results, '--', color='green')
decorate(xlabel = 'Year', ylabel = 'Monk Seal Population')
```
